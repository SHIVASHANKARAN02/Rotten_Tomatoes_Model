# DATA PREPROCESSING

```python
# Text Columns: Replace missing values with 'Unknown'
text_cols = ['movie_info', 'critics_consensus']
for col in text_cols:
    dataset[col] = dataset[col].fillna('Unknown')

# Categorical Columns: Replace missing values with 'Unknown'
categorical_cols = ['genre', 'directors', 'writers', 'cast', 'studio_name']
for col in categorical_cols:
    dataset[col] = dataset[col].fillna('Unknown')

# Numerical Columns: Replace missing values with the median
numerical_cols = ['runtime_in_minutes', 'audience_rating']
for col in numerical_cols:
    dataset[col] = dataset[col].fillna(dataset[col].median())

# Date Columns: Parse dates and replace missing with a default date
date_cols = ['in_theaters_date', 'on_streaming_date']
for col in date_cols:
    # Specify format explicitly
    dataset[col] = pd.to_datetime(dataset[col], format='%d-%m-%Y', errors='coerce')
    # Replace missing values with a default date
    dataset[col] = dataset[col].fillna(pd.Timestamp('1970-01-01'))

# Check the processed data
print(dataset)
```

# MODEL TRAINING AND EVALUATION

```python
import lightgbm as lgb
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Assuming the dataset is already preprocessed and in a pandas DataFrame
# Replace 'audience_rating' with your actual target column name
if 'audience_rating' not in dataset.columns:
    raise ValueError("Target column 'audience_rating' not found in the dataset.")

X = dataset.drop(columns=['audience_rating'])
Y = dataset['audience_rating']

# Convert datetime columns to numerical features (e.g., timestamps)
datetime_cols = X.select_dtypes(include=['datetime64']).columns.tolist()
if datetime_cols:
    print(f"Converting datetime columns: {datetime_cols}")
for col in datetime_cols:
    X[col] = X[col].astype('int64')  # Convert datetime to timestamp (int64)

# Identifying categorical columns for LightGBM
categorical_cols = X.select_dtypes(include=['object', 'category']).columns.tolist()
if categorical_cols:
    print(f"Identified categorical columns: {categorical_cols}")

# Convert categorical columns to category type
for col in categorical_cols:
    X[col] = X[col].astype('category')

# Split the data into training and validation sets
X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size=0.2, random_state=42)

# Create LightGBM Datasets
train_data = lgb.Dataset(X_train, label=Y_train, categorical_feature=categorical_cols, free_raw_data=False)
val_data = lgb.Dataset(X_val, label=Y_val, categorical_feature=categorical_cols, free_raw_data=False)

# Set parameters for LightGBM
params = {
    'objective': 'regression',  # Specify the regression task
    'metric': 'l2',  # Mean squared error (MSE)
    'learning_rate': 0.01,  # Learning rate
    'num_leaves': 31,  # The number of leaves in each tree
    'max_depth': 8,  # Max depth of each tree
    'boosting_type': 'gbdt',  # Gradient Boosting Decision Trees
    'bagging_fraction': 0.8,  # Random sampling of data for each iteration
    'feature_fraction': 0.8,  # Random sampling of features for each iteration
    'verbose': -1  # Suppress verbose output
}

# Train the model with early stopping
print("Starting LightGBM training...")
model = lgb.train(
    params,
    train_data,
    num_boost_round=2000,
    valid_sets=[train_data, val_data],
    valid_names=['train', 'validation'],
    callbacks=[lgb.early_stopping(stopping_rounds=200)]
)

# Predict on the validation data
y_pred = model.predict(X_val, num_iteration=model.best_iteration)

# Evaluate the model
mse = mean_squared_error(Y_val, y_pred)
r2 = r2_score(Y_val, y_pred)
print("\nEvaluation Metrics:")
print("Mean Squared Error (MSE):", mse)
print("R Score:", r2)

# Plot Feature Importance
print("\nPlotting feature importance...")
lgb.plot_importance(model, max_num_features=10, importance_type='split', title='Feature Importance')
plt.show()
```

# VISUALIZATION

```python
validation_results = pd.DataFrame({'Actual': Y_val, 'Predicted': y_pred})
print("\nActual vs Predicted Values (Sample):")
print(validation_results.head())

# Plot actual vs. predicted values
plt.figure(figsize=(8, 6))
plt.scatter(Y_val, y_pred, alpha=0.6, edgecolor='k', label='Predictions')
plt.plot([Y_val.min(), Y_val.max()], [Y_val.min(), Y_val.max()], color='red', linestyle='--', label='Ideal Fit')
plt.title('Actual vs Predicted Values')
plt.xlabel('Actual Audience Rating')
plt.ylabel('Predicted Audience Rating')
plt.legend()
plt.grid()
plt.show()
```
